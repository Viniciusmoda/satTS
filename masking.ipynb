```python
import gdal
```

```python
source = 'SMAP13_13500_14112_012_140814_L090_CX_01_hgt.tif'
ds = gdal.Open(source)

# get image size
rows = ds.RasterYSize
cols = ds.RasterXSize
bands = ds.RasterCount

# Set NoData Value
band = ds.GetRasterBand(1)
ndv = -10000.
band.SetNoDataValue(ndv)
```

```python
import matplotlib.pyplot as pyplot

values = band.ReadAsArray()
```

```python
pyplot.imshow(values)
```

```python
import numpy as np
import numpy.ma as ma

ndv = -10000.

data = values

masked_data = ma.masked_where(values == ndv, values)
print(masked_data)
```

```python
pyplot.imshow(masked_data)
```
